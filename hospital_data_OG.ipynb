```python
import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("healthdata.gov", None)


results = client.get("anag-cw7u", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
#on average beds available
hospital_data = results_df[['state','hospital_name','address','city','zip','total_beds_7_day_sum','inpatient_beds_used_7_day_sum','icu_beds_used_7_day_sum','total_icu_beds_7_day_sum',]]
```
